답변 :

- A/B test?
1. 두가지 버전을 비교해, 어떤게 더 효과저인지 비교하는 실험
2. 마케팅, 디자인 분야 등에서 광범위하게 사용되며, 클릭율이나 구매율 등을 근거로 무엇이 더 좋은 성과를 내는지 증명하는 과정
3. 감에 의존하지 않고 데이터 기반의 객관적 판단에 기여함


- 유의사항
1. 초두효과 : 사용자는 처음 접한 정보나 인터페이스를 강렬하게 기억하므로, 새로운 디자인에 불편함을 느껴 성과가 낮게 측정됨
2. 새로움효과 : 반대로 새로운 것에 호기심을 느껴 초반에 성과가 높게 측정됨
-> 사용자가 변화에 적응할 수 있도록 최소 1주 이상 진행하며, 데이터가 안정화되는 시점 확인
3. single variable : 요소를 동시에 바꾸면, 결과 비교시 어떤 요소가 원인인지 파악할 수 없음 
-> 여러 요소를 동시에 테스트 하고 싶다면, 다변량 테스트 사용
4. 충분한 크기의 표본을 확보하여 통계적 안정성을 확보해야함
5. 측정하려는 요소 외의 요인은 동일하게 통제해야함
6. 사용자 그룹 분할 시 무작위로 분산 시켜야 함

## Q2-2. 실험에 필요한 최소 test users 수를 구해봅시다.
유의수준 0.05, 검정력 80% 하에 수행해주세요.  


아래 공식 참고해서, 최소 test users 수를 구하는 코드 작성해주세요.

추가적으로 n은 산출하는데에 가정한 p1, p2도 같이 작성해주세요.

$$n = \frac{(Z_{\alpha/2} + Z_{\beta})^2 \cdot [p_1(1-p_1) + p_2(1-p_2)]}{(p_1 - p_2)^2}$$

## A2-2. 아래 질문에 대한 답과 코드를 채워주세요.

p1, p2는 각각 무엇을 의미하는 확률이고, 어떤 값으로 가정하셨나요?   
또한 위 공식을 보고 n을 구하기 위한 코드를 작성해주세요. 단, 라이브러리는 z값을 구하는 용도의 것만 사용해주세요.


### 답변:

```python
########

import scipy.stats as stats
import math

# 1. 파라미터 설정
p1 = 0.40  # 기존 "지금 인기" 배너 클릭률
p2 = 0.50  # 개선 후 목표 클릭률 (가설 기반 가정)
alpha = 0.05
power = 0.80

# 2. Z값 산출 (scipy.stats 활용)
z_alpha_half = stats.norm.ppf(1 - alpha/2) # 약 1.96
z_beta = stats.norm.ppf(power)             # 약 0.84

# 3. 공식 대입 (분자, 분모 계산)
numerator = (z_alpha_half + z_beta)**2 * (p1 * (1 - p1) + p2 * (1 - p2))
denominator = (p1 - p2)**2
n = numerator / denominator

# 4. 결과 출력
print(f"기존 클릭률(p1): {p1}")
print(f"목표 클릭률(p2): {p2}")
print(f"그룹당 필요한 최소 사용자 수(n): {math.ceil(n)}명")
print(f"총 필요 실험 사용자 수: {math.ceil(n) * 2}명")



#########
```

귀무가설 ($H_0$): 메인 홈 검색 모듈의 위치(하단 vs 최상단)와 상세 페이지 진입 여부 사이에는 상관관계가 없다. 즉, 두 그룹의 상세 페이지 진입률($CVR$)은 동일하다.$$H_0: p_A = p_B$$대립가설 ($H_1$): 메인 홈 검색 모듈을 최상단에 노출하는 것은 상세 페이지 진입 여부에 영향을 미친다. 즉, 실험군(B)의 상세 페이지 진입률이 대조군(A)보다 높을 것이다.$$H_1: p_A < p_B$$

## A3-1.  검정 이름과 이유를 설명해주세요.

답변: 위 가설에 필요한 검정은 카이제곱검정 이고,  

이유는 클릭함/클릭안함 으로 나뉘는 변수형 데이터 형태이고,파이썬과 같은 분석 도구 사용시 범주형이 처리하기에 더 용이하기 때문이다.

<br>

```python
'''''''''''''''''''''''''''''''''''''''
제가 세운 A/B TEST 설계 실험 코드입니다.
'''''''''''''''''''''''''''''''''''''''

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (실제 실험 결과 수치로 변경하세요)
# [상세진입 성공 수, 상세진입 실패 수]
control_results = [11500, 34500]  # 대조군(A)
variant_results = [12190, 33810]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 검색바 최상단 노출이 상세 페이지 진입률에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 검색바 위치 변경의 효과가 통계적으로 증명되지 않았습니다.")
```

## A3-4 아래 셀에 코드 작성해주세요. 그리고!!!
그리고 나온 값을 통해 수행하신 a/b test 결과를 설정하신 가설에 맞게 해석해주세요.

```python

import numpy as np
from scipy.stats import chi2_contingency

# 1. 실제 실험 데이터 입력 (예시 데이터)
# A: 기존 배너, B: 수정 배너
group_a_impressions = 385  # A그룹 총 노출수
group_a_clicks = 40       # A그룹 클릭수 (CTR 약 10.4%)

group_b_impressions = 385  # B그룹 총 노출수
group_b_clicks = 65       # B그룹 클릭수 (CTR 약 16.8%)

# 2. 분할표(Contingency Table) 생성
# [클릭한 유저 수, 클릭하지 않은 유저 수]
data = [
    [group_a_clicks, group_a_impressions - group_a_clicks],
    [group_b_clicks, group_b_impressions - group_b_clicks]
]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"--- A/B 테스트 검정 결과 ---")
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.6f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 배너 디자인 수정이 클릭률에 유의미한 차이를 만들어냈습니다! (성공)")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 두 배너 간의 클릭률 차이는 통계적으로 우연일 가능성이 높습니다.")

#############
```

**결과해석**
